```python
# Program courtesy of ChatGPT
x = np.random.rand(20)
y = np.random.rand(20)

# Calculate the best-fit line
slope, intercept, r_value, p_value, std_err = linregress(x, y)
best_fit_line = slope * x + intercept
r_squared = r_value**2

plt.figure(figsize=(8, 6))
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.scatter(x, y, color='lightblue', label='Random Points')
plt.plot(x, best_fit_line, color='blue', label=f'Best-fit line (y = {slope:.2f}x + {intercept:.2f}, $R^2$ = {r_squared:.2f})')
plt.xlabel('x')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter plot of random points with best-fit line')
plt.legend()
plt.grid(True)
plt.show()
```
